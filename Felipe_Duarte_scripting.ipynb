```python
import requests
from bs4 import BeautifulSoup

url = "https://pt.m.wikipedia.org/wiki/Lista_de_munic%C3%ADpios_de_Portugal_por_popula%C3%A7%C3%A3o"
data = requests.get(url).text
soup = BeautifulSoup(data, 'lxml')
```

```python

title = soup.title.string
title
print(title)
```

```python
tabela = soup.find('table', class_='wikitable')
data_rows = tabela.find_all('tr')
```

```python
rowsList = []
# percorrer as linhas
for row in data_rows:
    value = row.find_all('td')
    tdList = [elem.text.strip() for elem in value]
    # Ignora as colunas vazias
    if len(tdList) == 0:
        continue
    if len(tdList) < 5:
        regiao = tdList[3]
        tdList.append(regiao)
    tdList[2] = int(tdList[2].replace(u"\xa0", ""))    
    rowsList.append(tdList) 
rowsList 
```

```python
listaMunicipios = []
for municipio in rowsList:
    dict = {'Municipio': municipio[1], 'População': municipio[2], 'SubRegião': municipio[3], 'Região':municipio[4]}
    listaMunicipios.append(dict)
listaMunicipios
```

```python
def listaMunicipiosRegiao(regiao):
    for municipio in listaMunicipios:
        if municipio['Região']==regiao:
            print(municipio)
# listaMunicipiosRegiao('Norte')
```

```python
populacao = []
def listaMunicipiosPop():
    for municipio in listaMunicipios:
        populacao.append(municipio['População'])
    print ('total da população =', sum(populacao))
    print ('Média da população =', sum(populacao)/len(populacao))
    print ('Maior população =', min(populacao))
    print ('Menor população =', max(populacao))
listaMunicipiosPop()
```
