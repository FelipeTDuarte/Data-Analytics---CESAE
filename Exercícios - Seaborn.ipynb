Vamos trabalhar com um famoso conjunto de dados do titanic para estes exercícios:

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style('whitegrid')
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

## Exercícios
** Recrie os gráficos abaixo usando o dataframe titanic. Existem muito poucas dicas, pois a maioria dos gráficos pode ser feita com apenas uma ou duas linhas de código e uma dica basicamente revelaria a solução. Preste muita atenção aos rótulos x e y para obter dicas.**

* *Observação! Para não perder a imagem do gráfico, certifique-se de não codificar na célula que está diretamente acima do gráfico, há uma célula extra acima daquela que não substituirá o gráfico! **

```python
sns.jointplot(x='fare', y='age', data=titanic)
```

```python
sns.histplot(titanic['fare'], bins=30, kde=False, color='red')
```

```python
sns.boxplot(x='class', y='age', data=titanic, palette='rainbow')
```

```python
sns.swarmplot(x='class',y='age', data=titanic, palette='Set2', size=4)
```

```python
sns.countplot(x='sex', data=titanic)
```

```python
sns.heatmap(titanic.corr(), cmap='coolwarm')
plt.title('titanic.corr()')
```

```python
g = sns.FacetGrid(data=titanic, col='sex')
g.map(plt.hist,'age')
```
