
1. Elaborar programa em Python com o intuito de gestão de uma loja online, concretizando os pontos em baixo mencionados.
	a. Ações:
		i. Efetuar Login
			1. Efetuar login como cliente, como funcionário e como administrador.
			2. Caso não tenha login, efetuar registo de cliente.
		ii. Efetuar compra
			1. Selecionar o produto e quantidade do produto com stock.
			2. Checkout
			3. Imprimir resumo da compra no ecrã. (talão)
		iii. Criação de compra para stock
			1. Criação de artigo novo
			2. Adicionar stock a artigo existente.
		iv. Criação de Relatórios:
			1. Relatório de stock (stock total)
			2. Relatório de vendas por produto (Utilizador indica qual produto pretende ver as vendas)
	b. Estruturas:
		i. O stock deve incluir id do produto, quantidade, nome do produto, preço custo (O preço de venda é + 30%)
		ii. O "talão" deve ter a seguinte estrutura.
			1. Número da fatura
			2. Número do cliente
			3. Nome do produto
			4. Quantidade
			5. Preço s/IVA
			6. IVA
			7. Total C/iva (23%)
		iii. Todas as vendas devem ser armazenadas em base de dados (Id da venda, nome do cliente, id do produto, quantidade comprada, total da venda)
		iv. Dados do utilizador: NIF, Nome, telefone, localidade.

```python
#create the tables of the Data Base
# Connect to the database (or create it if it doesn't exist)
conn = sqlite3.connect('database.db')
cursor = conn.cursor()

# Create employee table
cursor.execute('''
    CREATE TABLE IF NOT EXISTS employee_table (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        username TEXT NOT NULL,
        password TEXT NOT NULL,
        is_admin BOOLEAN NOT NULL
    )
''')

# Create client table
cursor.execute('''
    CREATE TABLE IF NOT EXISTS client_table (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        username TEXT NOT NULL,
        password TEXT NOT NULL,
        name TEXT NOT NULL,
        nif TEXT NOT NULL,
        phone TEXT NOT NULL,
        location TEXT NOT NULL

)
''')

# Create product table
cursor.execute('''
    CREATE TABLE IF NOT EXISTS product_table (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        name TEXT NOT NULL,
        price REAL NOT NULL,
        stock INTEGER NOT NULL
    )
''')

# Create sale table
cursor.execute('''
    CREATE TABLE IF NOT EXISTS sale_table (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        client_name INTEGER NOT NULL,
        product_id INTEGER NOT NULL,
        quantity INTEGER NOT NULL,
        total REAL NOT NULL,
        FOREIGN KEY(client_name) REFERENCES client_table(name),
        FOREIGN KEY(product_id) REFERENCES product_table(id)
    )
''')

# Commit changes and close connection
conn.commit()
conn.close()
```

```python
# Function to read and insert data from CSV file to database
def insert_data_from_csv(file_name, table_name, is_admin = False):
    with open(file_name, 'r') as file:
        csvreader = csv.reader(file)
        for row in csvreader:
            if table_name == "employee_table":
                if is_admin:
                    cursor.execute("SELECT * FROM employee_table WHERE username = ?", (row[0],))
                    if cursor.fetchone() is None:
                        cursor.execute("INSERT INTO employee_table (username, password, is_admin) VALUES (?, ?, ?)", (row[0], row[1], 1))
                else:
                    cursor.execute("SELECT * FROM employee_table WHERE username = ?", (row[0],))
                    if cursor.fetchone() is None:
                        cursor.execute("INSERT INTO employee_table (username, password, is_admin) VALUES (?, ?, ?)", (row[0], row[1], 0))
            elif table_name == "client_table":
                cursor.execute("SELECT * FROM client_table WHERE username = ?", (row[0],))
                if cursor.fetchone() is None:
                    cursor.execute("INSERT INTO client_table (username, password, name, nif, phone, location) VALUES (?, ?, ?, ?, ?, ?)", (row[0], row[1], row[2], row[3], row[4], row[5]))
            elif table_name == "product_table":
                cursor.execute("SELECT * FROM product_table WHERE name = ?", (row[1],))
                if cursor.fetchone() is None:
                    cursor.execute("INSERT INTO product_table (name, price, stock) VALUES (?, ?, ?)", (row[1], row[2], row[3]))
            elif table_name == "sale_table":
                cursor.execute("SELECT * FROM sale_table WHERE client_name = ? AND product_id = ?", (row[1], row[2]))
                if cursor.fetchone() is None:
                    cursor.execute("INSERT INTO sale_table (client_name, product_id, quantity, total) VALUES (?, ?, ?, ?)", (row[1], row[2], row[3], row[4]))
            else:
                print(f"Invalid table name: {table_name}")
```

```python
# Function to save data from database
def save_to_csv(table_name):
    conn = sqlite3.connect('database.db')
    cursor = conn.cursor()
    cursor.execute(f"SELECT * FROM {table_name}")
    data = cursor.fetchall()
    with open(f"{table_name}.csv", 'w', newline='') as file:
        csvwriter = csv.writer(file)
        csvwriter.writerow([i[0] for i in cursor.description])
        csvwriter.writerows(data)
    #print(f"{table_name} exported to {table_name}.csv")
    conn.close()
```

```python
#function of clients actions       
def client_function(client_id):
    # Connect to the database
    conn = sqlite3.connect('database.db')
    cursor = conn.cursor()
    #revenue
    r=0.3
    #VAT
    vat=0.23
    
    
    # Print available products
    cursor.execute("SELECT * FROM product_table")
    products = cursor.fetchall()
    print("\n Available products:")
    for product in products:
        print(f"{product[0]}. {product[1]} - Stock: {product[3]} - Price: {product[2]*(1+r+vat):.2f}")

    # Ask user to select product and quantity
    selected_product_id = int(input("Select product by ID: "))
    selected_quantity = int(input("Select quantity: "))

    # Check if selected product is in stock
    cursor.execute(f"SELECT stock FROM product_table WHERE id = {selected_product_id}")
    product_stock = cursor.fetchone()[0]
    if selected_quantity > product_stock:
        print("\n Not enough stock.\n")
        return
    
    # Get product details
    cursor.execute(f"SELECT name, price FROM product_table WHERE id = {selected_product_id}")
    product_name, product_price = cursor.fetchone()
    # Get sale id
    cursor.execute("SELECT max(id) FROM sale_table")
    sale_id = cursor.fetchone()[0]
        
    # Checkout
    print(f"Product: {product_name}")
    print(f"Quantity: {selected_quantity}")
    print(f"Price: {product_price*(1+r+vat)*selected_quantity:.2f}")
    print(f"VAT: {product_price*(1+r+vat)-product_price:.2f}")
    total_price = product_price*(1+r+vat)*selected_quantity
    confirm = input("Confirm purchase? (y/n)")
    if confirm == 'y':
        cursor.execute(f"UPDATE product_table SET stock = stock - {selected_quantity} WHERE id = {selected_product_id}")
        cursor.execute(f"INSERT INTO sale_table (client_name, product_id, quantity, total) VALUES (?, ?, ?, ?)", (client_id, selected_product_id, selected_quantity, total_price))
        conn.commit()
        
        
        # Write admin data to CSV file
        invoice_id = cursor.lastrowid
        with open(f'invoice_{sale_id}.csv', mode='w') as invoice_file:
            invoice_writer = csv.writer(invoice_file)
            #invoice_writer.writerow(["Invoice Number","Client Number","Product Name", "Quantity", "Price w/o VAT", "VAT (23%)", "Total (w/ VAT)"])
            #invoice_writer.writerow([invoice_id, client_id, product_name, selected_quantity, product_price*(1+r), product_price*(1+r) * vat, total_price])
        
            invoice_writer.writerow([f"Invoice Number: {invoice_id}"])
            invoice_writer.writerow([f"Client Number: {client_id}"])
            invoice_writer.writerow([f"Product: {product_name}"])
            invoice_writer.writerow([f"Quantity: {selected_quantity}"])
            invoice_writer.writerow([f"Price w/o VAT: {product_price*(1+r)}"])
            invoice_writer.writerow([f"VAT (23%): {product_price*(1+r) * vat}"])
            invoice_writer.writerow([f"T]otal (w/ VAT): {total_price}"])
        
        print('\n Invoice:')
        print(f"Invoice Number: {invoice_id}")
        print(f"Client Number: {client_id}")
        print(f"Product: {product_name}")
        print(f"Quantity: {selected_quantity}")
        print(f"Prce w/o VAT: {product_price*(1+r)}")
        print(f"VAT (23%): {product_price*(1+r) * vat}")
        print(f"T]otal (w/ VAT): {total_price} \n")
        
        print("Purchase completed successfully.")
    else:
        print("Purchase canceled.")

    add_more = input("Do you want to add more products? (y/n)")
    if add_more == "y":
        client_function(client_id)
    else:
        return 



# Close connection
conn.close()
```

```python
def admin_function():
    # Connect to the database
    conn = sqlite3.connect('database.db')
    cursor = conn.cursor()
    while True:
        # Print options
        print("1. Create new employee")
        print("2. Create new admin")
        print("3. View stock report")
        print("4. View sales report")
        print("5. Exit")

        # Get user input
        option = int(input("Select option: "))

        # Create new employee
        if option == 1:
            employee_username = input("Employee username: ")
            employee_password = input("Employee password: ")
            cursor.execute("INSERT INTO employee_table (username, password, is_admin) VALUES (?, ?, ?)", (employee_username, employee_password, 0))

        # Create new admin
        elif option == 2:
            admin_username = input("Admin username: ")
            admin_password = input("Admin password: ")
            cursor.execute("INSERT INTO employee_table (username, password, is_admin) VALUES (?, ?, ?)", (admin_username, admin_password, 1))

        # View stock report
        elif option == 3:
            cursor.execute("SELECT * FROM product_table")
            products = cursor.fetchall()
            for product in products:
                print(f"{product[1]} - Stock: {product[3]} - Price: {product[2]}")

        # View sales report
        elif option == 4:
            product_name = input("Enter product name: ")
            cursor.execute(f"SELECT COUNT(product_id), SUM(quantity) FROM sale_table WHERE product_id = (SELECT id FROM product_table WHERE name = '{product_name}')")
            sales_data = cursor.fetchone()
            print(f"\n Total sales of {product_name}: {sales_data[0]}")
            print(f"Total quantity sold: {sales_data[1]} \n")

        # Commit changes and close connection
            conn.commit()
        elif option == 5:
            break
        
        conn.close()
```

```python
def employee_function():
    conn = sqlite3.connect('database.db')
    cursor = conn.cursor()
    while True:
        print("1. Add new product")
        print("2. Add stock to existing product")
        print("3. Remove product")
        print("4. Update product")
        print("5. Exit")
        action = input("Enter your choice: ")
        if action == '1':
            add_new_product(cursor)
            conn.commit()
            print("\n Product added successfully.\n")
        elif action == '2':
            product_name = input("Enter product name: ")
            stock_quantity = input("Enter stock quantity: ")
            cursor.execute("UPDATE product_table SET stock = stock + ? WHERE name = ?", (stock_quantity, product_id))
            conn.commit()
            print("\n Stock added successfully.\n")
        elif action == '3':
            remove_product()
        elif action == '4':
            update_product()
        elif action == '5':
            break
        else:
            print('\n Invalid choice. \n')
    conn.close()

def add_new_product(cursor):
    while True:
        product_name = input("Enter product name: ")
        cursor.execute("SELECT name FROM product_table")
        existing_products = cursor.fetchall()
        existing_product_names = [p[0] for p in existing_products]
        if product_name in existing_product_names:
            print("\n Product already exists. Please enter a different product name. \n")
            continue
        product_price = input("Enter product price: ")
        product_stock = input("Enter product stock: ")
        cursor.execute("INSERT INTO product_table (name, price, stock) VALUES (?, ?, ?)", (product_name, product_price,product_stock))
        print("\n Product added successfully.\n")
        break
    

def remove_product():
    conn = sqlite3.connect('database.db')
    cursor = conn.cursor()
    product_id = int(input("Enter product id: "))
    cursor.execute("SELECT id, name FROM product_table")
    products = cursor.fetchall()
    product_ids = [p[0] for p in products]
    if product_id not in product_ids:
        print("\n Invalid product id. \n")
        return
    selected_product = None
    for product in products:
        if product[0] == int(product_id):
            selected_product = product
            break
    if selected_product:
        confirm = input(f"Are you sure you want to remove the product {selected_product[1]}? (y/n)")
        if confirm.lower() == "y":
            cursor.execute("DELETE FROM product_table WHERE id = ?", (product_id,))
            conn.commit()
            print(f"Product {selected_product[1]} removed successfully.")
        else:
            print("\n Product removal canceled. \n")
    else:
        print("\n Invalid product id. \n")
    conn.close()


def update_product():
    conn = sqlite3.connect('database.db')
    cursor = conn.cursor()
    product_id = int(input("Enter product id: "))
    cursor.execute("SELECT id FROM product_table")
    products = cursor.fetchall()
    product_ids = [p[0] for p in products]
    if product_id not in product_ids:
        print("Invalid product id.")
        return
    print("1. Update product name")
    print("2. Update product price")
    print("3. Update product stock")
    choice = input("Enter your choice: ")
    if choice == '1':
        new_name = input("Enter new product name: ")
        cursor.execute("UPDATE product_table SET name = ? WHERE id = ?", (new_name, product_id))
    elif choice == '2':
        new_price = input("Enter new product price: ")
        try:
            new_price = float(new_price)
        except ValueError:
            print("Invalid price.")
            return
        cursor.execute("UPDATE product_table SET price = ? WHERE id = ?", (new_price, product_id))
    elif choice == '3':
        new_stock = input("Enter new stock quantity: ")
        try:
            new_stock = int(new_stock)
        except ValueError:
            print("Invalid stock quantity.")
            return
        cursor.execute("UPDATE product_table SET stock = ? WHERE id = ?", (new_stock, product_id))
    conn.commit()
    print("\n Product updated successfully. \n")
    conn.close()
    
def register():
    # Collect user data
    email = input("Enter email: ")
    if not validate_email.validate_email(email):
        print("Invalid email address.")
        return
    password = input("Enter your desired password: ")
    name = input("Enter your name: ")
    nif = input("Enter NIF: ")
    if not re.match("^[0-9]{9}$", nif):
        print("Invalid NIF.")
    phone = input("Enter your phone number: ")
    location = input("Enter your address: ")

    # Connect to the database
    conn = sqlite3.connect('database.db')
    cursor = conn.cursor()

    # Insert user data into client table
    cursor.execute("INSERT INTO client_table (email, password, name, nif, phone, location) VALUES (?, ?, ?, ?, ?, ?)", (email, password, name, nif, phone, location))
    conn.commit()
    
    # update csv file
    with open("client.csv", "w", newline='') as file:
        writer = csv.writer(file)
        cursor.execute("SELECT * FROM client_table")
        rows = cursor.fetchall()
        for row in rows:
            writer.writerow(row)
    print("\n Registration successful! Please login to continue. \n")
    conn.close()
```
