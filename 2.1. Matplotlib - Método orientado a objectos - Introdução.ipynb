** Importe matplotlib.pyplot como plt e defina %matplotlib inline. Qual é o comando usado se não estiver a usar o jupyter notebook?**

```python
import numpy as np
x = np.arange(0,100)
y = x * 2
z = x ** 2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
#plt.show()
```

### Exercício 1
Siga estes passos:  <br>

 * Crie um objeto figure chamado fig usando plt.figure()  <br>
 * Use add_axes para adicionar um eixo à tela da figura em [0,0,1,1]. Chame esse novo eixo de ax.  <br>
 * Plot (x,y) nesses eixos e defina os rótulos e títulos para corresponder ao gráfico abaixo: <br>

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')
```

### Exercício 2
 Crie um objeto figure e coloque dois eixos, ax1 e ax2. Localizado em [0,0,1,1] e [0.2,0.5,.2,.2] respectivamente.

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.2,.2])
```

 Agora plot (x,y) em ambos os eixos. E chame o objeto figure para mostrá-lo.

```python
ax1.plot(x,y,'b')
ax1.set_xlabel('x')
ax1.set_ylabel('y')

ax2.plot(x,y, 'b')
ax2.set_xlabel('x')
ax2.set_ylabel('y')

fig
```
