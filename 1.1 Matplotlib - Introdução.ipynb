4. Escreva um programa Python para traçar duas ou mais linhas no mesmo gráfico com legendas adequadas para cada linha. <br>
Deverá ter um output semelhante ao seguinte: <br>

```python
x1 = [10,20,30]
y1 = [20,40,10]
plt.plot(x1,y1, label = "line 1")
x2 = [10,20,30]
y2 = [40,10,30]
plt.plot(x2,y2, label = "line 2")
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.title('Duas ou mais linhas no mesmo plot com legendas')
plt.legend()
```

5. Crie um gráfico semelhante ao anterior com legendas, larguras e cores diferentes. <br>
Deverá ter um output semelhante ao seguinte: <br>

```python
x1 = [10,20,30]
y1 = [20,40,10]
plt.plot(x1,y1, color = 'blue', linewidth = 3, label = "line1-width-3")
x2 = [10,20,30]
y2 = [40,10,30]
plt.plot(x2,y2, color = 'red', linewidth = 5, label = "line2-width-5")
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.title('Duas ou mais linhas no mesmo plot com legendas')
plt.legend()
```

6. Crie um gráfico semelhante ao anterior com legendas, larguras e cores diferentes. <br>
Deverá ter um output semelhante ao seguinte: <br>

```python
x1 = [10,20,30]
y1 = [20,40,10]
plt.plot(x1,y1, color = 'blue', linewidth = 3, label = "line1-width-3", linestyle = 'dotted')
x2 = [10,20,30]
y2 = [40,10,30]
plt.plot(x2,y2, color = 'red', linewidth = 5, label = "line2-width-5", linestyle = 'dashed')
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.title('Duas ou mais linhas no mesmo plot com legendas')
plt.legend()
```

7. Crie um gráfico com duas ou mais linhas e defina os marcadores de linha.<br>
Deverá ter um output semelhante ao seguinte:<br>

```python
x = [1,4,5,6,7]
y = [2,6,3,6,3]
plt.plot(x,y, color = 'red', linestyle = 'dashdot', linewidth = 3, marker = 'o', markerfacecolor = 'blue', markersize = 12)
plt.ylim(1,8)
plt.xlim(1,8)
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.title('Mostrar markers')
```
