### Exercício 2
Mostre um gráfico de barras (bar chart) da popularidade das seguintes linguages de programação. <br>
x = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++'] <br>
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7] <br>
x_pos = [i for i, _ in enumerate(x)] <br>

```python
x = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
x_pos = [i for i, _ in enumerate(x)]

plt.bar(x_pos, popularity, color='blue')
plt.xlabel("Linguagens")
plt.ylabel("Popularidade")
plt.title("Popularidade das linguagens de programação\n" + "Em todo o mundo, Oct 2017 comparado com o ano anterior")
plt.xticks(x_pos, x)
# Ligar a grelha
plt.minorticks_on()
plt.grid(which='major', linestyle='-', linewidth='0.5', color='red')
# Ajuste da grelha menor
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.xlim([-0.4, len(x)+0.5])
plt.ylim([0, max(popularity)+0.5])
plt.show()
```

# Exercício 6
Crie um gráfico de pizza das conquistas de medalhas de ouro dos cinco países mais bem-sucedidos nos Jogos Olímpicos de 2016. Leia os dados de um arquivo ex6.csv.

```python
df =  pd.read_csv('ex6.csv')
country_data = df["country"]
medal_data = df["gold_medal"]
colors = ["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728", "#8c564b"]
explode = (0.1, 0, 0, 0, 0)  
plt.pie(medal_data, labels=country_data, explode=explode, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)
plt.title("Medalhas de ouro dos 5 melhores\n"+" países nas olimpiadas de verão de 2016")
plt.show()
```
